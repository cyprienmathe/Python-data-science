```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import requests

# URL avec les spécifications
api_url = "https://equipements.sports.gouv.fr/api/explore/v2.1/catalog/datasets/data-es/records?select=COUNT(*)&where=dep_nom%20is%20not%20null&group_by=dep_nom"
# Conversion en dictionnaire json
req = requests.get(api_url)
wb = req.json()

# Création de la dataframe

df = pd.DataFrame(wb['results'])
df_sorted = df.sort_values(by='COUNT(*)', ascending=True)
```

```python
#Graphe test
plt.figure(figsize=(10,6))
plt.bar(df_sorted["dep_nom"], df_sorted["COUNT(*)"])
```
